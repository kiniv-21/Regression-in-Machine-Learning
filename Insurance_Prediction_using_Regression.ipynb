```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('insurance.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.replace({'sex':{'male':0,'female':1}},inplace = True)
df.head()
```

```python
df.replace({'smoker':{'yes':0,'no':1}},inplace = True)
df.head()
```

```python
df.replace({'region':{'southwest':0,'southeast':1,'northwest':2,'northeast':3}},inplace = True)
df.head()
```

```python
X = df.drop(columns=['charges'],axis=1)
y = df['charges']
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
dtr = DecisionTreeRegressor()
```

```python
dtr.fit(X_train,y_train)
```

```python
rfr = RandomForestRegressor()
rfr.fit(X_train,y_train)
```

```python
lr_pred = lr.predict(X_test)
dtr_pred = dtr.predict(X_test)
rfr_pred = rfr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
print("Linear Regression: ",r2_score(y_test,lr_pred))
```

```python
print("Decsion Tree: ",r2_score(y_test,dtr_pred))
```

```python
print("Random Forest: ",r2_score(y_test,rfr_pred))
```
